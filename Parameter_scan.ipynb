```python
## Inputs
# Input video path
inPath = 'OneDrive_1_02-07-2021/MVI_0425.mp4'
dataloc = '/home/sultannazir/Videos/' 
## Parameter values
# Threshold values
Tvals = [15,20,25,30,35,40,45,50]
# Blob size ranges
blob_sizes = [[0.01,10]]
size_range_key = ['All']
# Number of individuals 
N = 15
```

```python
# Generate PV files for all parameter combinations
for T in Tvals:
    key=0
    for size in blob_sizes:
        op = 'T{}_{}'.format(T,size_range_key[key])        # name of PV file for current parameters
        loc = dataloc+"data/"+op                     # path for saving tracking data
        f= open(op+".settings","w+")                 # edit settings file with TRex parameters
        f.write("threshold = "+str(T)+"\n"+
               "blob_size_range = "+str(size)+"\n"+
               "meta_real_width = "+str(80)+"\n"+
               "output_csv_decimals = 2 \n"+
               "output_format = csv \n"+
               "output_interpolate_positions = false \n"+
               "fishdata_dir = "+loc+"\n"+
               'output_graphs = [["X",["RAW","WCENTROID"]],["Y",["RAW","WCENTROID"]],["X",["RAW","HEAD"]],["Y",["RAW","HEAD"]],'+
               '["VX",["RAW","HEAD"]],["VY",["RAW","HEAD"]],["AX",["RAW","HEAD"]],["AY",["RAW","HEAD"]],["ANGLE",["RAW"]],'+
               '["segment_length",["RAW"]],["SPEED",["RAW","WCENTROID"]],["time",[]],["frame",[]],["num_pixels",[]]] \n'+
               "track_max_individuals = "+str(N)+"\n"+
               "track_threshold = "+str(T))
        f.close()
        settings = dataloc+op+".settings"
        # command line
        !export MESA_LOADER_DRIVER_OVERRIDE=i965; tgrabs -i $inPath -o $op -settings $settings -average_samples 200 -averaging_method max

        # command line 
        !mkdir $loc
        !export MESA_LOADER_DRIVER_OVERRIDE=i965;trex -i $op -auto_quit true
        key+=1
```

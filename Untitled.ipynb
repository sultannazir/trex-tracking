```python
import cv2
import seaborn as sns
import numpy as np
from PIL import Image
from jupyterthemes import jtplot
jtplot.style(theme='monokai', context='notebook', ticks=True, grid=False)
```

```python
# input RAW file to extract frame
cam = cv2.VideoCapture('OneDrive_1_02-07-2021/MVI_0425.mp4')
W = cam.get(cv2.CAP_PROP_FRAME_WIDTH)
H = cam.get(cv2.CAP_PROP_FRAME_HEIGHT)
```

```python
cam.set(cv2.CAP_PROP_POS_FRAMES,frameno-1)
ret, frame = cam.read()
# convert to grayscale
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
```

```python
# grayscale 2D histogram of selected frame
sns.heatmap(gray[400:770,1021:1249], vmin=0, vmax=200, cmap = 'Greys_r', cbar_kws={'label': 'f(x,y)'})
```

```python
# grayscale 2D histogram of averaged BG with max method
BG = np.array(Image.open('average_T25_All.png').convert('L'))
sns.heatmap(BG[400:770,1021:1249], vmin=0, vmax=200,cmap='Greys_r', cbar_kws={'label': 'b(x,y)'})
```

```python
# grayscale 2D histogram of averaged BG with mode method
BG = np.array(Image.open('average_MVI_0425.png').convert('L'))
sns.heatmap(BG, vmin=0, vmax=200, cbar_kws={'label': 'b(x,y)'})
```

```python
# background subtraction
max_diff = 100
Z = np.zeros((int(H),int(W)))
for i in range(int(H)):
    for j in range(int(W)):
        Z[i][j] = abs(int(BG[i][j])-int(gray[i][j])) # avoid subtraction of grayscale values because 
                                                        #negative values of t are set to 255+t
sns.heatmap(Z,  vmin=threshold, vmax=max_diff,cbar_kws={'label': 't(x,y) = b(x,y) - f(x,y)'})
```

```python
# zoom in with pixel coordinates to see fish with background subtraction
zoom = Z
sns.heatmap(zoom[400:770,1021:1249],  vmin=0, vmax=200, cmap="Greys_r", cbar_kws={'label': 'd(x,y) = b(x,y) - f(x,y)'})
```

```python
# blobs
binary = zoom>34
c = binary.astype(int)
sns.heatmap(c[400:770,1021:1249],cmap='Greys_r', cbar_kws={'label': 't(x,y)'})
```
